```python
#EASY TASK 1:

n = int(input())

for i in range(n):
    
    inp = input().split(' ')
    
    for i in range(0, len(inp)):
    
        for index in range(0, len(inp)-i-1):
        
            if (len(inp[index]) < len(inp[index+1])):
            
                temp = inp[index]
            
                inp[index] = inp[index+1]
            
                inp[index+1] = temp
            
    ans = ' '.join(inp)

    print(ans)
```

```python
#EASY TASK 2:

n = int(input())

even_list = []

odd_list = []

for i in range(n):
    
    inp = int(input())
    
    if (inp % 2 == 0):

        even_list.append(inp)
      
    else:
        
        odd_list.append(inp)

for i in sorted(even_list):

  print(i)

for i in sorted(odd_list, reverse=True):

  print(i)
```

```python
#EASY TASK 3:

n = int(input())

for i in range(n):

  data = input().split(' ')

  num_1 = data[0]

  num_2 = data[1]

  ans = ''

  if (len(num_1) == len(num_2)):
    
    for index in range(len(num_1)):

        ans = ans + num_1[index] + num_2[index]
    
  elif (len(num_1) > len(num_2)):
    
    extra = num_1[len(num_2):]
    
    for index in range(len(num_2)):
        
      ans = ans + num_1[index] + num_2[index]
        
    ans = ans + extra
       
  else:
      
    extra = num_2[len(num_1):]
    
    for index in range(len(num_1)):
        
      ans = ans + num_1[index] + num_2[index]
        
    ans = ans + extra
    
  print(ans)
```

```python
#MEDIUM TASK 1:

n = int(input())

for i in range(n):

  list1 = []

  inp = input()

  for num in inp:
    
    list1.append(int(num))

  for index1 in range(len(list1)-1):
    
    min_val = list1[index1]
    
    min_loc = index1

    for index2 in range(index1+1, len(list1)):

      if (index1 == 0 and list1[index2] == 0):

        pass

      else:
        
        if (list1[index2] < min_val):
            
            min_val = list1[index2]
            
            min_loc = index2

    temp = min_val
    
    list1[min_loc] = list1[index1]
    
    list1[index1] = temp
    
  lucky_num = ''

  for num in list1:

    lucky_num += str(num)

  print(lucky_num)
```

```python
#MEDIUM TASK 3:

n = int(input())

for i in range(n):

  list1 = []

  count = 0

  inp = input()

  for num in inp:
    
    if (ord(num) == 32):

      list1.append(num)

    else:

      if (count % 2 == 0):

        list1.append(num.upper())

      else:

        list1.append(num.lower())

      count +=1

  ans = ''.join(list1)

  print(ans)
```
