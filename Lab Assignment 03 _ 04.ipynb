```python
#Task 2:

class Shape:

  def __init__(self, name, x, y):

    self.name = name

    self.x = x

    self.y = y

  def area(self):

    if self.name == 'Triangle':

      area = 0.5 * self.x * self.y

    elif self.name == 'Rectangle':

      area = self.x * self.y

    elif self.name == 'Rhombus':

      area = 0.5 * self.x * self.y

    elif self.name == 'Square':

      area = self.x * self.y

    else:

      area = 'Shape Unknown'

    print('Area:', area)


triangle = Shape("Triangle",10,25)
triangle.area()
print("==========================")
square = Shape("Square",10,10)
square.area()
print("==========================")
rhombus = Shape("Rhombus",18,25)
rhombus.area()
print("==========================")
rectangle = Shape("Rectangle",15,30)
rectangle.area()
print("==========================")
trapezium = Shape("Trapezium",15,30)
trapezium.area()
```

```python
#Task 3:

class Calculator:

  def __init__(self):

    print('Calculator is ready!')

  def calculate(self, x, y, op):

    self.x = x

    self.y = y

    self.op = op

    self.ans = 0

    if self.op == '+':

      self.ans = self.x + self.y

    elif self.op == '-':

      self.ans = self.x - self.y

    elif self.op == '*':

      self.ans = self.x * self.y

    else:

      self.ans = self.x / self.y

    return self.ans

  def showCalculation(self):

    print(self.x, self.op, self.y, '=', self.ans)


c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()
```

```python
#Task 9:

class Batsman:

  def __init__(self, *args):

    if len(args) == 2:
      
      self.name = 'New Batsman'

      self.run = args[0]

      self.ball = args[1]

    else:

      self.name = args[0]

      self.run = args[1]

      self.ball = args[2]

  def printCareerStatistics(self):

    print('Name:', self.name, end = '\n')

    print(f'Runs Scored: {self.run} , Balls Faced: {self.ball}')

  def battingStrikeRate(self):

    return (self.run / self.ball) * 100

  def setName(self, name):

    self.name = name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#Task 11:

class TaxiLagbe:

  def __init__(self, info, area):

    self.info = info 

    self.area = area

    self.pas_list = []

    self.fare = 0

  def addPassenger(self, *args):

    for i in args:

      if len(self.pas_list) <= 3:

        name, fare = i.split('_')

        self.pas_list.append(name)

        self.fare += int(fare)

        print(f'Dear {name}! Welcome to TaxiLagbe.')

      else:

        print('Taxi Full! No more passengers can be added.')

        break

  def printDetails(self):

    print('Trip info for Taxi number:', self.info)

    print(f'This taxi can cover only {self.area} area.')

    print('Total passengers:', len(self.pas_list))

    print('Passenger lists:')

    print(', '.join(self.pas_list))

    print(f'Total collected fare: {self.fare} Taka')


taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#Task 13:

class StudentDatabase:

  def __init__(self, name, id):

    self.name = name

    self.id = id

    self.grades = {}

  def calculateGPA(self, course, semester):

    c_list = []

    gpa_list = []

    s = {}

    for i in course:

      c, gpa = str(i).split(':')

      c_list.append(c)

      gpa_list.append(gpa)

      gpa = 0

      for i in gpa_list:

        gpa += float(i)

        final_gpa = gpa / len(gpa_list)

    s[tuple(c_list)] = final_gpa

    self.grades[semester] = s

  def printDetails(self):

    print('Name:', self.name)

    print('ID:', self.id)

    for key, var in self.grades.items():

      print(f'Courses taken in {key}:')

      for keyy, varr in var.items():

        for k in keyy:

          print(k)

        print('GPA:', varr)


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
