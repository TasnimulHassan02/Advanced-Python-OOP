```python
#Dictionary & Tuple:-----------------------

#Task1:

user1 = input('Please enter the first dictionary: ').replace(' ', '')

user2 = input('Please enter the first dictionary: ').replace(' ', '')

n1 = user1.split(',')

n2 = user2.split(',')

dic1 = {}

dic2 = {}

for i in n1:
    
    j = i.split(':')
    
    dic1[j[0]] = int(j[1])

for i in n2:
    
    j = i.split(':')
    
    dic2[j[0]] = int(j[1])

final_dic = {}

for key1 in dic1.keys():
    
    for key2 in dic2.keys():
        
        if key1 == key2:
            
            final_dic[key2] = dic1[key1] + dic2[key2]
            
    if key1 not in final_dic:
        
        final_dic[key1] = dic1[key1]
        
    if key2 not in final_dic:
        
        final_dic[key2] = dic2[key2]
        
print(final_dic)

list1 = []

for value in final_dic.values():
    
    if value not in list1:
        
        list1.append(value)
    
list1.sort()

print('Values:', tuple(list1))
```

```python
#Funcion:-----------------------

#Task1:

def BMI(height, weight):
    
    height = height / 100
    
    bmi = (weight / height**2)
    
    print('Your score is', bmi, end = '. ')
    
    if bmi < 18.5:
        
        print('You are Underweight')
        
    elif (18.5 <= bmi <= 24.9):
        
        print('You are Normal')
        
    elif (25 <= bmi <= 30):
        
        print('You are Overweight')
        
    else:
        
        print('You are Obese')
    
n = input('Please enter your height and weight: ').split(',')

for i in range(len(n)):
    
    n[i] = int(n[i])
    
a_tuple = tuple(n)

(h, w) = a_tuple

BMI(h, w)
```

```python
#Task2:

def div_add(minm, maxm, div):
    
    count = 0
    
    for num in range(minm, maxm):
        
        if (num % div == 0):
            
            count += num
            
    return count

n = input('Please enter the values: ').split(',')

for i in range(len(n)):
    
    n[i] = int(n[i])
    
a_tuple = tuple(n)

(minm, maxm, div) = a_tuple

print(div_add(minm, maxm, div))
```

```python
#Task4:

def replace_domain(email, new_dom, old_dom = 'kaaj.com'):
    
    eml = email.split('@')
    
    name = eml[0]
    
    dom = eml[1]
    
    if dom == new_dom:
        
        print('Unchanged:', email)
        
    else:
        
        new_eml = name + '@' + new_dom
        
        print('Changed:', new_eml)
        
n = input('Please enter the inputs: ').split(', ')

a_tuple = tuple(n)

if len(a_tuple) == 2:
    
    (email, new_dom) = a_tuple
    
else:
    
    (email, new_dom, old_dom) = a_tuple
    
replace_domain(email, new_dom, old_dom)
```

```python
#Task5:

def palindrome_checker(sent):
    
    old_sent = sent.replace(' ','')
    
    new_sent = ''
    
    for i in range(-1, -len(old_sent)-1, -1):
        
        new_sent = new_sent + old_sent[i]
        
    if new_sent == old_sent:
        
        print('Palindrome')
        
    else:
        
        print('Not a Palindrome')
        
n = input('Please enter the sentence: ')

palindrome_checker(n)
```

```python
#Task6:

def sent_cap(sent):
    
    d = ['.', '!', '?']
    
    sent[0] = sent[0].capitalize()
    
    for i in range(len(sent)):
        
        if list(sent[i])[-1] in d:
            
            try:
                
                sent[i+1] = sent[i+1].capitalize()
            
            except:
                
                pass
                
        if sent[i]  == 'i':
            
            sent[i] = sent[i].capitalize()
            
    return ' '.join(sent)
    
n = list(input('Please enter your paragraph: ').split())

print(sent_cap(n))
```
